```python
import pandas as pd
import numpy as np
def delUnname0(df):
    df = df.drop('Unnamed: 0',axis=1)
    return df
Customer_Flow = pd.read_csv("Customer_Flow.csv")
Customer_Flow = delUnname0(Customer_Flow)

shop_id_1 =Customer_Flow[Customer_Flow["shop_id"] == 1] 
```

```python
shop_id_1['data'] = pd.to_datetime(shop_id_1['data'])
shop_id_1.index = shop_id_1['data']
shop_id_1.drop(['shop_id','data'],axis=1,inplace=True)
```

```python
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = pd.rolling_mean(timeseries, window=14)
    rolstd = pd.rolling_std(timeseries, window=14)

    #Plot rolling statistics:
    orig = timeseries.plot(kind='line',figsize=[24,8],color='blue',label='Original',title = 'orig')
    mean = rolmean.plot(kind='line',figsize=[24,8],color='red', label='Rolling Mean',title = 'mean')
    std = rolstd.plot(kind='line',figsize=[24,8],color='black', label = 'Rolling Std',title = 'std')
    plt.show()
    
    #Perform Dickey-Fuller test:
    print 'Results of Dickey-Fuller Test:'
    
    a_timeseries = np.array(timeseries)
    dim_1_data = []
    for x in a_timeseries:
        dim_1_data.extend(x)
        
    dftest = adfuller(dim_1_data, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print dfoutput
```

```python
# 从原序列可以看出是稳定序列
# 检验统计量比5%的临界值小，所以我们在95%的置信区间认为它是稳定序列。但是值相差比较小。区分度不大。
test_stationarity(shop_id_1)
```

```python
# 这里取log 取对数变换相当于增加了一个惩罚机制，数据越大惩罚越大！！
ts_log = np.log(shop_id_1)
```

```python
moving_avg = pd.rolling_mean(ts_log,31)
moving_avg.plot(kind='line',title='moving avg 31',figsize=[12,8])
plt.show()
```

```python
ts_log_moving_avg_diff  = ts_log - moving_avg
# 原来序列减去取log和31天滑动平均后发现
# 检验统计量比5%的临界值小很多（相对于原序列的稳定性）
# 比1%还小，这比以前的情况还好
# 所以我们在95%的置信区间认为它是稳定序列。
# Test Statistic                  -4.542797
# Critical Value (5%)             -2.869833
ts_log_moving_avg_diff.dropna(inplace=True)
test_stationarity(ts_log_moving_avg_diff)
```

```python
# 取“加权移动平均法”可以对最近的值赋予更高的权重。   
# 尴尬的是，虽然序列平滑了很多，但是却不稳定了
# Test Statistic                  -2.462234
# Critical Value (5%)             -2.869341
expwighted_avg = pd.ewma(ts_log, halflife=31)
ts_expwighted_avg  = ts_log - expwighted_avg
ts_expwighted_avg.dropna(inplace=True)
test_stationarity(ts_expwighted_avg)
```

```python
# 既然一个月的加权平均不好就尝试2周 取两周的发现效果还不错，比取一个月的还好
# Test Statistic                  -4.605966
# Critical Value (5%)             -2.869176
# 请注意在这种情况下就不会有遗漏值因为所有的值在一开始就被赋予了权重。
# 所以在运行的时候，它没有先前的值参与。
expwighted_avg = pd.ewma(ts_log, halflife=14)
ts_expwighted_avg  = ts_log - expwighted_avg
ts_expwighted_avg.dropna(inplace=True)
test_stationarity(ts_expwighted_avg)
```

```python
# 使用一阶差分来消除趋势 这里的稳定性明显比刚才的要好很多很多
# DF检验统计量小于1% 的临界值，因此该时间序列在90%的置信区间上是稳定的。
# 我们同样可以采取二阶或三阶差分在具体应用中获得更好的结果
# Test Statistic                -7.627035e+00
# Critical Value (1%)           -3.448052e+00
ts_log_diff = ts_log - ts_log.shift()
ts_log_diff.dropna(inplace=True)
test_stationarity(ts_log_diff)
```

```python
# 这里尝试做一阶的复原 假设预测的是ts_log_diff 目标恢复成ts_log,实际上恢复的方法是把第一个元素作为一个基值，然后加上残差值
ts_log_diff_cumsum = ts_log_diff.cumsum()
base_value = ts_log.ix[0]
a = []
b = base_value.values
for x in range(1,len(ts_log)+1):
    a.append(b)
predictions_ARIMA_log = pd.DataFrame(a, index=ts_log.index)
predictions_ARIMA_log.rename(columns={0:'Num'},inplace=True)
predictions_ARIMA_log = predictions_ARIMA_log.add(ts_log_diff_cumsum,fill_value=0)

ts_log = ts_log.ix[predictions_ARIMA_log.index]
rmse = np.sqrt(np.sum((predictions_ARIMA_log-ts_log)**2)/ts_log.size)
```

```python
# 使用二阶的时候，Test Statistic就有微弱的上升，当采用三阶的时候上升很多
# 和上面做对比,采用一阶会好很多
ts_log_diff = ts_log - ts_log.shift().shift()
ts_log_diff.dropna(inplace=True)
test_stationarity(ts_log_diff)
```

```python
# 这里则使用分解的办法
from statsmodels.tsa.seasonal import seasonal_decompose
# 感情这里的freq是要填数字而不是‘D’，我特么的。。。。
# 这里的频率指的是天数（相隔多少天）
# 要指出的是，取出差不多一个月的Trend已经比较平滑了，14天的也还可以
decomposition = seasonal_decompose(ts_log,freq=31)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.subplot(411)
plt.plot(ts_log, label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal,label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
# 检查残差的稳定性
ts_log_decompose = residual
ts_log_decompose.dropna(inplace=True)
test_stationarity(ts_log_decompose)
# Test Statistic                -8.163490e+00
# Critical Value (1%)           -3.449392e+00
# 这已经很稳定了。我滴天。
# 你应该注意到, 在这种情况下将残差转换为原始值对未来数据不是很直观。
```
