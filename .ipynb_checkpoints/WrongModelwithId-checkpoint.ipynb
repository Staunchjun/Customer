```python
import pandas as pd
import xgboost as xgb
import numpy as np
def delUnname0(df):
    df = df.drop('Unnamed: 0',axis=1)
    return df
Customer_Flow = pd.read_csv("Customer_Flow.csv")
Customer_Flow = delUnname0(Customer_Flow)
```

```python
# ramdomly choose two shop.Reference ARMA/AR/MA/XGboost lr
import matpotlib.pyplot as plt
shop_id_1 =Customer_Flow[Customer_Flow["shop_id"] == 1] 
shop_id_1000 =Customer_Flow[Customer_Flow["shop_id"] == 1000] 
shop_id_1.plot(kind='line')
shop_id_1000.plot(kind='line')
plt.show()
```

```python
Customer_Flow['data'] = pd.to_datetime(Customer_Flow['data'])
```

```python
def xgb_eval_custom_r(y_pred, dtrain):
    y_true = dtrain.get_label()
    up = np.sum(y_pred - y_true)
    down = np.sum(y_pred + y_true)
    whole = np.abs(up/down)
    Loss = whole/len(y_true)
    #here only predict one shop 
    return 'Loss', Loss
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
#preprocessing normailze?No.Do a one hot encoding~~Because the data is so small~~
def TimeFeat(df):
    columns = df.columns
    df['month'] = pd.Index(df[columns[0]]).month
    df['day'] = pd.Index(df[columns[0]]).day
    df['dayofweek'] = pd.Index(df[columns[0]]).dayofweek
    dummies_month = pd.get_dummies(df['month'],prefix='month')
    dummies_day = pd.get_dummies(df['day'],prefix='day')
    dummies_dayofweek = pd.get_dummies(df['dayofweek'],prefix='dayofweek')  
    df_dummies = df.join([dummies_month,dummies_day,dummies_dayofweek],how='outer')
    df_dummies = df_dummies.drop(['month','day','dayofweek'],axis=1)
    return df_dummies
```

```python
startt = pd.to_datetime("2016-10-18")
endt = pd.to_datetime("2016-10-31")
```

```python
p_startt = pd.to_datetime("2016-11-01")
p_endt = pd.to_datetime("2016-11-14")
p_time =pd.date_range(start=p_startt,end=p_endt,freq='D')
```

```python
p_shop_id = range(1,2001,1)
```

```python
shopid = []
p_t = []
for shop_id in p_shop_id:
    for t in p_time:
        shopid.append(shop_id)
        p_t.append(t)
predict ={'data':p_t,'shop_id':shopid}
predict = pd.DataFrame(predict)
```

```python
copy_Customer_Flow = Customer_Flow.drop(['Num'],axis=1)
```

```python
all_data  = copy_Customer_Flow.merge(predict,how='outer')
```

```python
all_data = TimeFeat(all_data)
```

```python
test_df =all_data[((all_data['data'] >= startt) & (all_data['data'] <= endt))]
train_df =all_data[(all_data['data'] < startt)]
wLabel_test =Customer_Flow[((Customer_Flow['data'] >= startt) & (Customer_Flow['data'] <= endt))]
wLabel_train =Customer_Flow[(Customer_Flow['data'] < startt)]
test_df = test_df.merge(wLabel_test,how='outer')
train_df = train_df.merge(wLabel_train,how='outer')
predict_df =all_data[((all_data['data'] >= p_startt) & (all_data['data'] <= p_endt))]
```

```python
pd.set_option('display.max_columns',120)
```

```python
ToDrop = ['Num','data']
train_xgb = xgb.DMatrix(data=train_df.drop(ToDrop,axis=1)
                        ,label=train_df['Num'])
valid_xgb = xgb.DMatrix(data=test_df.drop(ToDrop,axis=1)
                        ,label=test_df['Num'])                        
params = {
  'objective': 'reg:linear'
  ,'eta': 0.1
  ,'max_depth': 10
  , 'subsample': 0.8
  , 'colsample_bytree': 0.6
  ,'min_child_weight': 2
  ,'gamma': 0
  , 'seed': 10
,'reg_alpha': 0.7
}       
evallist = [(train_xgb, 'train'), (valid_xgb, 'valid')]
model = xgb.train(params.items()
                  , dtrain=train_xgb
                  , num_boost_round=10000
                  , evals=evallist
                  , early_stopping_rounds=20
                  , maximize=False
                  , verbose_eval=10
#                   , feval=xgb_eval_custom_r
              )

print ('get info from model')                  
print (model.best_score, model.best_iteration,model.best_ntree_limit)
```

```python
xgb.plot_importance(model)
plt.show()
```

```python
predict_df = predict_df.drop(['data'],axis=1)
```

```python
predict_df = pd.DataFrame(predict_df)
# train_df.drop(ToDrop,axis=1)
```

```python
predict_df_xgb = xgb.DMatrix(data=predict_df)
```

```python
result = model.predict(predict_df_xgb,ntree_limit=model.best_ntree_limit)
```

```python
result = pd.DataFrame(result)
```

```python
result = result.rename(columns={0:'Num'})
```

```python
prediction = predict.join(result,how='outer')
```

```python
p_result = []
for shop_id in p_shop_id:
    each_line = []
    each_line.append(shop_id)
    eachShop = prediction[prediction['shop_id'] == shop_id]
    for Num in eachShop.Num:
        each_line.append(Num) 
    p_result.append(each_line)
```

```python
p_result = pd.DataFrame(p_result)
```

```python
p_result.to_csv('prediction.csv',header=False,index=False,encoding='utf-8')
```
